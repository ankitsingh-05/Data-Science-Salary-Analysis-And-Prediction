```python
df['Salary Estimate'].value_counts()
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df['Salary Estimate'].value_counts()
```

```python
salary = df['Salary Estimate'].apply(lambda x : x.split('(')[0])
```

```python
salary.unique()
```

```python
salary
```

```python
salary = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
salary
```

```python
salary.values
```

```python
df['Hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df.drop(['emploer_provided'],axis=1,inplace=True)
df.head()
```

```python
salary = salary.apply(lambda x: x.lower().replace('per hour', ''))
salary = salary.apply(lambda x: x.lower().replace('employer provided salary', ''))
```

```python
salary.values
```

```python
df['salary'] = salary
```

```python
salary = salary.apply(lambda x: x.replace(':',''))
```

```python
# df['min_salary'] = salary.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = salary.apply(lambda x: int(x.split('-')[1]))
```

```python
 df['min_salary'] = salary.apply(lambda x: int(x.split('-')[0]))
```

```python
salary.values
```

```python
df.head()
```

```python
df['avg_salary'] = (df['max_salary']+df['min_salary'])/2
```

```python
df
```

### company name

```python
df['Company'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3], axis=1)
```

```python
df.head()
```

### location & headquarters 

```python
df['state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df.state.value_counts()
```

```python
df['at_headquarter'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

```python
df.dtypes
```

```python
df['Age'] = df.Founded.apply(lambda x: x if x<1 else 2023-x)
```

```python
df.head()
```

```python
#parsing of job description (python, etc.)

#python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
 
#r studio 
df['R'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)

#spark 
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)

#aws 
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)

#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.to_csv('salary_data.csv',index=False)
```
